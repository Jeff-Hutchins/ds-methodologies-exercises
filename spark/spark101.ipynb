```python
mpg = data('mpg')
```

```python
mpg.head()
```

```python
mpg = spark.createDataFrame(mpg)
```

```python
mpg.show(5)
```

```python
mpg.printSchema()
```

```python
mpg.count(), len(mpg.columns)
```

```python
mpg.select(concat(
    lit('The')
    col('year')
    lit(' ')
    col('manufacturer')
    lit(' ')
    col('model')
    lit(' has a ')
    col('cyl')
    lit(' cylinder engine'),
).alias('vehicle_cylinder_description')).show(truncate=False)
```
